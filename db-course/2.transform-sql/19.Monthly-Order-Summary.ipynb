```python
%sql
CREATE TABLE IF NOT EXISTS gizmobox.gold.order_monthly_summary
AS
Select
  date_format(transaction_timestamp, 'yyyy-MM') AS order_month,
  customer_id,
  count(distinct order_id) as total_orders,
  SUM(quantity) as total_items,
  SUM(price* quantity) as total_amont 
from
  gizmobox.silver.orders
GROUP by
order_month,customer_id;
```
