```python
%sql
CREATE TABLE IF NOT EXISTS gizmobox.gold.customers_address
AS
SELECT
  c.customer_id,
  c.customer_name,
  c.date_of_birth,
  c.email,
  c.member_since,
  c.telephone,
  a.billing_address_line_1,
  a.billing_city,
  a.billing_state,
  a.billing_postcode,
  a.shipping_address_line_1,
  a.shipping_city,
  a.shipping_state,
  a.shipping_postcode
from
  gizmobox.silver.customers c
    JOIN gizmobox.silver.addresses a
      ON c.customer_id = a.customer_id;
```
