```python
%python
dbutils.widgets.text("container_name", "demo")
dbutils.widgets.text("storage_name", "deacoursestoragewestus2")
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS demo.delta_lake
  MANAGED LOCATION 'abfss://${container_name}@${storage_name}.dfs.core.windows.net/delta_lake'
```

```python
%sql
CREATE TABLE IF NOT EXISTS demo.delta_lake.companies(
  company_name STRING,
  founded_date DATE,
  country STRING
);
```

```python
%sql
DESCRIBE EXTENDED demo.delta_lake.companies;
```
