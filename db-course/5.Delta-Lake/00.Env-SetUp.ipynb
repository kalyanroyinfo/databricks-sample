```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS dea_course_ext_dl_demo
    URL 'abfss://${container_name}@${storage_name}.dfs.core.windows.net'
    WITH (STORAGE CREDENTIAL `dea_course_ext_sc`)
    COMMENT 'External Location'
```
