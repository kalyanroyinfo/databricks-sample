### Behaviour of DROP and CREATE STATEMENTS

```python
%sql
DROP TABLE IF EXISTS demo.delta_lake.companies1;
CREATE TABLE IF NOT EXISTS demo.delta_lake.companies1(
  company_name STRING,
  founded_date DATE,
  country STRING)
  COMMENT 'This table contain information about some of the successful companies'
  TBLPROPERTIES (sensitive = 'true');

INSERT into demo.delta_lake.companies1 (company_name, founded_date, country)
  VALUES
    ("test2", "1974-11-11", "abcd1"),
    ("test3", "1975-11-11", "abcd2"),
    ("test4", "1984-11-11", "abcd3");
```

```python
%sql
DESC HISTORY demo.delta_lake.companies1;
```

### Behaviour of CREATE and REPLACE STATEMENTS

```python
%sql
DROP TABLE IF EXISTS demo.delta_lake.companies1;
```

```python
%sql

CREATE OR REPLACE TABLE demo.delta_lake.companies1(
  company_name STRING,
  founded_date DATE,
  country STRING)
  COMMENT 'This table contain information about some of the successful companies'
  TBLPROPERTIES (sensitive = 'true');

INSERT into demo.delta_lake.companies1 (company_name, founded_date, country)
  VALUES
    ("test2", "1974-11-11", "abcd1"),
    ("test3", "1975-11-11", "abcd2"),
    ("test4", "1984-11-11", "abcd3");
```

```python
%sql
DESC HISTORY demo.delta_lake.companies1;
```

### CTAS Statement
####Column Data Type inferred we cannot set.
####To Change the column ttype CAST() the data type.

```python
%sql
DROP TABLE demo.delta_lake.companies_india;
```

```python
%sql
CREATE TABLE demo.delta_lake.companies_india
AS
Select
  *
from
  demo.delta_lake.companies
where
  country = 'India';
```

```python
%sql
SELECT * from demo.delta_lake.companies_india;
```

```python
%sql
DESC HISTORY demo.delta_lake.companies_india;
```

```python
%sql
ALTER TABLE demo.delta_lake.companies_india ALTER COLUMN founded_date COMMENT 'THIS IS COMMENT'
```

```python
%sql
DESC EXTENDED demo.delta_lake.companies_india;
```
