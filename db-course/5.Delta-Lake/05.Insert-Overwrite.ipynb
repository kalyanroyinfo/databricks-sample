### Insert OverWrite
- ### Replace All the Data in a table
- ### Replace all the data from a specific partition
- ###How to Handle Schema Change


- **INSERT OVERWRITE** : Overwrite the existing data in a table or a spesific partition with the new data.
- **INSERT INTO** : APPEND New Data 

```python
%sql
DROP TABLE IF EXISTS demo.delta_lake.gold_companies;
CREATE TABLE IF NOT EXISTS demo.delta_lake.gold_companies(
  company_name STRING,
  founded_date DATE,
  country STRING);

INSERT into demo.delta_lake.gold_companies (company_name, founded_date, country)
  VALUES
    ("test2", "1974-11-11", "India"),
    ("test3", "1975-11-11", "US");
```

```python
%sql
SELECT * FROM demo.delta_lake.gold_companies;
```

```python
%sql
DROP TABLE IF EXISTS demo.delta_lake.bronze_companies;
CREATE TABLE IF NOT EXISTS demo.delta_lake.bronze_companies(
  company_name STRING,
  founded_date DATE,
  country STRING);

INSERT into demo.delta_lake.bronze_companies (company_name, founded_date, country)
  VALUES
    ("test2", "1974-11-11", "India"),
    ("test3", "1975-11-11", "US"),
    ("test44", "1964-11-11", "India"),
    ("test5", "1955-11-11", "US"),
    ("test6", "1984-11-11", "China");
```

```python
%sql
SELECT * from demo.delta_lake.bronze_companies;
```

```python
%sql
INSERT OVERWRITE TABLE demo.delta_lake.gold_companies
SELECT * FROM demo.delta_lake.bronze_companies;
```

```python
%sql
SELECT * FROM demo.delta_lake.gold_companies;
```

```python
%sql
DESC HISTORY demo.delta_lake.gold_companies;
```

```python
%sql
DROP TABLE IF EXISTS demo.delta_lake.gold_companies_partition;
CREATE TABLE IF NOT EXISTS demo.delta_lake.gold_companies_partition(
  company_name STRING,
  founded_date DATE,
  country STRING)
  PARTITIONED BY (country);

INSERT into demo.delta_lake.gold_companies_partition (company_name, founded_date, country)
  VALUES
    ("test2", "1974-11-11", "India"),
    ("test3", "1975-11-11", "US");
```

```python
%sql
DESC EXTENDED demo.delta_lake.gold_companies_partition
```

```python
%sql
SELECT * FROM demo.delta_lake.gold_companies_partition;
```

```python
%sql
DROP TABLE IF EXISTS demo.delta_lake.bronze_companies_usa;
CREATE TABLE IF NOT EXISTS demo.delta_lake.bronze_companies_usa(
  company_name STRING,
  founded_date DATE,
  country STRING);

INSERT into demo.delta_lake.bronze_companies_usa (company_name, founded_date, country)
  VALUES
    ("test2", "1974-11-11", "US"),
    ("test3", "1975-11-11", "US"),
    ("test44", "1964-11-11", "US"),
    ("test5", "1955-11-11", "US"),
    ("test6", "1984-11-11", "US");

SELECT * FROM demo.delta_lake.bronze_companies_usa;
```

```python
%sql
SELECT * FROM demo.delta_lake.bronze_companies_usa;
```

```python
%sql
INSERT OVERWRITE TABLE demo.delta_lake.gold_companies_partition
PARTITION (country="US")
SELECT company_name, founded_date FROM demo.delta_lake.bronze_companies_usa;
```

```python
%sql
SELECT * FROM demo.delta_lake.gold_companies_partition;
```
