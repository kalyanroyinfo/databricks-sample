```python
%sql

USE CATALOG gizmobox;

CREATE SCHEMA  IF NOT EXISTS  landing
      MANAGED LOCATION 'abfss://${container_name}@${storage_name}.dfs.core.windows.net/landing'
      COMMENT 'This is landing schema for gizmobox';

CREATE SCHEMA  IF NOT EXISTS  bronze
      MANAGED LOCATION 'abfss://${container_name}@${storage_name}.dfs.core.windows.net/bronze'
      COMMENT 'This is landing schema for gizmobox';

CREATE SCHEMA  IF NOT EXISTS  silver
      MANAGED LOCATION 'abfss://${container_name}@${storage_name}.dfs.core.windows.net/silver'
      COMMENT 'This is landing schema for gizmobox';

CREATE SCHEMA  IF NOT EXISTS  gold
      MANAGED LOCATION 'abfss://${container_name}@${storage_name}.dfs.core.windows.net/gold'
      COMMENT 'This is landing schema for gizmobox';
```

```python
%sql
SHOW SCHEMAS;
```

```python
%sql
USE CATALOG gizmobox;
USE SCHEMA landing;

CREATE EXTERNAL VOLUME IF NOT EXISTS  operational_data
     LOCATION 'abfss://${container_name}@${storage_name}.dfs.core.windows.net/landing/operational_data/';

CREATE EXTERNAL VOLUME IF NOT EXISTS  external_data
     LOCATION 'abfss://${container_name}@${storage_name}.dfs.core.windows.net/landing/external_data/';
```
