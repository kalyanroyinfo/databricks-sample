### regexp_replace importent function for exam

```python
%sql
CREATE OR REPLACE TEMP VIEW tv_orders_fixed
AS
SELECT 
    value,
    regexp_replace(value, '"order_date": (\\d{4}-\\d{2}-\\d{2})', '"order_date": "\$1"') AS fixed_value 
FROM gizmobox.bronze.v_orders;
```

```python
%sql
select * from tv_orders_fixed;
```

### schema_of_json() in Databricks

```python
%sql
select schema_of_json(fixed_value) AS schema,fixed_value from tv_orders_fixed LIMIT 1;
```

### from_json in databricks

```python
%sql
select from_json(fixed_value,"STRUCT<customer_id: BIGINT, items: ARRAY<STRUCT<category: STRING, details: STRUCT<brand: STRING, color: STRING>, item_id: BIGINT, name: STRING, price: BIGINT, quantity: BIGINT>>, order_date: STRING, order_id: BIGINT, order_status: STRING, payment_method: STRING, total_amount: BIGINT, transaction_timestamp: STRING>") AS json_value,fixed_value from tv_orders_fixed;
```

```python
%sql
DROP TABLE gizmobox.silver.orders_json;
```

```python
%sql
CREATE TABLE IF NOT EXISTS gizmobox.silver.orders_json
AS
select from_json(fixed_value,"STRUCT<customer_id: BIGINT, items: ARRAY<STRUCT<category: STRING, details: STRUCT<brand: STRING, color: STRING>, item_id: BIGINT, name: STRING, price: BIGINT, quantity: BIGINT>>, order_date: STRING, order_id: BIGINT, order_status: STRING, payment_method: STRING, total_amount: BIGINT, transaction_timestamp: STRING>") AS json_value from tv_orders_fixed;
```

```python
%sql
select * from gizmobox.silver.orders_json;
```
