### Commonly Used Higher Order Array Functions:
- TRANSFORM
- FILTER
- EXISTS
- AGGREGATE

### SYNTAX:
function_name (array_column, lambda_expression)

lambda_expression: element -> expression

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW order_items as
SELECT * FROM
VALUES
(1, array('smartphone','laptop','monitor')),
(2, array('tablet','headephopne','smartwatch')),
(3, array('keyboard','mouse'))
AS orders(order_id,items);
```

```python
%sql
SELECT * FROM order_items;
```

```python
%sql
SELECT
  order_id,
  transform(items, x -> UPPER(x)) as upper_items
from
  order_items;
```

```python
%sql
SELECT
  order_id,
  FILTER(items, x -> x LIKE '%smart%') as smart_items
from
  order_items;
```

# Higher Order Function MAP

### _Commonly Used Higher Order MAP Function_
- TRANSFORM_VALUE
- TRANSFORM_KEYS
- MAP FILTER

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW order_items_prices as
SELECT * FROM
VALUES
(1, map('smartphone',699,'laptop',111,'monitor',599)),
(2, map('tablet',599,'headephopne',199,'smartwatch',299)),
(3, map('keyboard',89,'mouse',59))
AS orders(order_id,item_prices);
```

```python
%sql
select * from order_items_prices;
```

```python
%sql
select order_id,
      TRANSFORM_KEYS(item_prices,(k,v)->upper(k)) as upper_item_prices
         from order_items_prices;|
```

```python
%sql
select order_id,
      TRANSFORM_VALUES(item_prices,(k,v)->ROUND(v*1.10, 2)) as price_with_ta
         from order_items_prices;
```

```python
%sql
select order_id,
      MAP_FILTER(item_prices,(k,v)->v>500) as price_with_500
         from order_items_prices;
```
