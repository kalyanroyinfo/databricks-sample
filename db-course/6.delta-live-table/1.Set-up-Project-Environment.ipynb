####1. Create External Location
#Location Name: dea_course_ext_dl_circuitbox
#ADLS Path abfss://circuitbox@deacoursestoragewestus22.dfs.core.windows.net/
#Storage Credential: dea_course_ext_sc

```python
%sql
CREATE EXTERNAL LOCATION IF NOT EXISTS dea_course_ext_dl_circuitbox
    URL 'abfss://circuitbox@deacoursestoragewestus22.dfs.core.windows.net'
    WITH (STORAGE CREDENTIAL `dea_course_ext_sc`)
    COMMENT 'External Location for the circuitbox storage account'
```

```python
%fs ls 'abfss://circuitbox@deacoursestoragewestus22.dfs.core.windows.net/'
```

- Create Catalog: circuitbox
- Managed Location: abfss://circuitbox@deacoursestoragewestus22.dfs.core.windows.net/

```python
%sql
CREATE CATALOG IF NOT EXISTS circuitbox
MANAGED LOCATION 'abfss://circuitbox@deacoursestoragewestus22.dfs.core.windows.net/'
COMMENT 'catalog for the circuitbox data lakehouse'
```

```python
%sql
SHOW CATALOGS;
```

```python
%sql
USE CATALOG circuitbox;

CREATE SCHEMA IF NOT EXISTS landing
MANAGED LOCATION 'abfss://circuitbox@deacoursestoragewestus22.dfs.core.windows.net/landing';

CREATE SCHEMA IF NOT EXISTS lakehouse
MANAGED LOCATION 'abfss://circuitbox@deacoursestoragewestus22.dfs.core.windows.net/lakehouse';
```

```python
%sql
SHOW SCHEMAS;
```

```python
%sql
USE CATALOG circuitbox;
USE SCHEMA landing;

create external volume if not exists operational_data
location 'abfss://circuitbox@deacoursestoragewestus22.dfs.core.windows.net/landing/operational_data'
```
