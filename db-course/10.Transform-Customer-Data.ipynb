```python
%sql
SELECT
  DISTINCT *
FROM
  gizmobox.bronze.v_customers
where
  customer_id IS NOT NULL
ORDER BY
  customer_id;
```

```python
%sql
SELECT
  customer_id,
  MAX(customer_name),
  MAX(date_of_birth),
  MAX(email),
  MAX(member_since),
  MAX(created_timestamp),
  MAX(telephone)
FROM
  gizmobox.bronze.v_customers
where
  customer_id IS NOT NULL
GROUP BY customer_id
ORDER BY
  customer_id;
```

```python
%sql
CREATE OR REPLACE TEMPORARY VIEW v_customers_distinct 
AS
SELECT
  DISTINCT *
FROM
  gizmobox.bronze.v_customers
where
  customer_id IS NOT NULL
ORDER BY
  customer_id;
```

```python
%sql
SELECT
  customer_id,
  MAX(created_timestamp) as max_created_timestamp
FROM
  v_customers_distinct
GROUP BY
  customer_id;
```

```python
%sql
WITH cte_max AS (
  SELECT
    customer_id,
    MAX(created_timestamp) as max_created_timestamp
  FROM
    v_customers_distinct
  GROUP BY
    customer_id
)
SELECT
  *
from
  v_customers_distinct t
    JOIN cte_max c
      ON c.customer_id = t.customer_id
      and c.max_created_timestamp = t.created_timestamp
```

```python
%sql
WITH cte_max AS (
  SELECT
    customer_id,
    MAX(created_timestamp) as max_created_timestamp
  FROM
    v_customers_distinct
  GROUP BY
    customer_id
)
SELECT
  CAST(t.created_timestamp AS TIMESTAMP)  AS created_timestamp,
  t.customer_id,
  t.customer_name,
  CAST(t.date_of_birth AS DATE) AS date_of_birth,
  t.email,
  CAST(t.member_since AS DATE ) AS member_since,
  t.telephone
from
  v_customers_distinct t
    JOIN cte_max c
      ON c.customer_id = t.customer_id
      and c.max_created_timestamp = t.created_timestamp
```

```python
%sql
CREATE TABLE gizmobox.silver.customers
AS
WITH cte_max AS (
  SELECT
    customer_id,
    MAX(created_timestamp) as max_created_timestamp
  FROM
    v_customers_distinct
  GROUP BY
    customer_id
)
SELECT
  CAST(t.created_timestamp AS TIMESTAMP)  AS created_timestamp,
  t.customer_id,
  t.customer_name,
  CAST(t.date_of_birth AS DATE) AS date_of_birth,
  t.email,
  CAST(t.member_since AS DATE ) AS member_since,
  t.telephone
from
  v_customers_distinct t
    JOIN cte_max c
      ON c.customer_id = t.customer_id
      and c.max_created_timestamp = t.created_timestamp
```
